# How does store size relate to weekly sales and markdowns?

```python
store_data = combined_df.groupby('Store')
store_data
```

```python
# this line of code is creating a summarized version of the combined_df DataFrame, 
# where the mean values of selected columns (Weekly_Sales, store size and 5 different markdowns) are calculated for each store. 
# This is useful for analyzing the average weekly sales, store size, and markdown values across different stores

store_data = combined_df.groupby('Store').agg({'Weekly_Sales':'mean', 'Size':'mean', 
                                      'MarkDown1':'mean', 'MarkDown2':'mean',
                                      'MarkDown3':'mean', 'MarkDown4':'mean',
                                      'MarkDown5':'mean'}).reset_index()

```

```python
# Creating a column for total markdowns
store_data['Total_MarkDown'] = store_data[['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']].sum(axis=1)

store_data.head()
```

# sales at different markdowns

```python

# Group data by markdowns and calculate mean sales
sales_by_markdown1 = df.groupby('MarkDown1')['Weekly_Sales'].mean().reset_index()
sales_by_markdown2 = df.groupby('MarkDown2')['Weekly_Sales'].mean().reset_index()
sales_by_markdown3 = df.groupby('MarkDown3')['Weekly_Sales'].mean().reset_index()
sales_by_markdown4 = df.groupby('MarkDown4')['Weekly_Sales'].mean().reset_index()
sales_by_markdown5 = df.groupby('MarkDown5')['Weekly_Sales'].mean().reset_index()

# Visualize the data
plt.figure(figsize=(10, 6))
sns.lineplot(data=sales_by_markdown1, x='MarkDown1', y='Weekly_Sales')
sns.lineplot(data=sales_by_markdown2, x='MarkDown2', y='Weekly_Sales')
sns.lineplot(data=sales_by_markdown3, x='MarkDown3', y='Weekly_Sales')
sns.lineplot(data=sales_by_markdown4, x='MarkDown4', y='Weekly_Sales')
sns.lineplot(data=sales_by_markdown5, x='MarkDown5', y='Weekly_Sales')
plt.title('Weekly Sales (mean) by Markdown')
plt.xlabel('Markdown')
plt.ylabel('Mean Weekly Sales')
plt.xticks(rotation=45)
plt.show()
```

the plot says there is a linear relation ship between sales and markdown

```python
# Plotting the data
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Size", y="Total_MarkDown", data=store_data, size="Weekly_Sales", sizes=(20, 200))
plt.title('Total Markdowns vs Store Size')
plt.xlabel('Store Size')
plt.ylabel('Total Markdowns')
plt.show()
```

This scatter plot illustrates the relationship between store size and total markdowns with each point scaled by average weekly sales. As we can see, larger stores tend to have more markdowns. One might assume this is due to larger stores having a wider variety of products and a higher sales volume which could lead to more promotional activities.

Moreover the size of the points represents the average weekly sales and it is noticeable that larger stores, which also have more markdowns, tend to have higher sales. This obviously suggests that markdowns are an effective strategy for boosting sales particularly in larger stores.

What isn't obvious is that while one might expect larger stores to have better inventory management and thus fewer markdowns, the data suggests the opposite. This could lead to other interesting ideas for insights such as looking into the effectiveness of markdowns on sales across different store sizes.

Can we predict future sales based on historical sales data and other features?

```python
combined_df.columns
```

```python
df_forecast = combined_df.drop(['Date','Weekly_Sales', 'IsHoliday', 'Temperature', 'Year', 'CPI'], axis = 1)
df_forecast.columns
```

```python
df_forecast.info()
```

```python
# Perform one-hot encoding
one_hot_encoded = pd.get_dummies(df_forecast['Type'])

# Concatenate the one-hot encoded DataFrame with the original DataFrame
# axis=1 concatenates along columns
df_encoded = pd.concat([df, one_hot_encoded], axis=1)

# Drop the original categorical column if desired
# df_encoded.drop('Category', axis=1, inplace=True)

# Display the resulting DataFrame
# print(df_encoded)
df_encoded.columns
```

```python
df_encoded.shape
```

```python
df_encoded.isna().sum()
```

```python
df_encoded.fillna(0, inplace=True)
```

```python
df_encoded.isna().sum()
```

```python
df_encoded.columns
```

```python
X = df_encoded.drop(['Weekly_Sales'], axis = 1)
y = df_encoded['Weekly_Sales']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
print(y_test.tail(1000))
```

```python
Linear Regression is suitbale when a continuous variable is a re
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

```python
y_pred = linreg.predict(X_test)
```

```python
# measures the average squared difference between the actual and predicted values. 
# Lower values of MSE indicate better performance.
# R^2 score ranges from 0 to 1, where 1 indicates a perfect fit and 
# 0 indicates no relationship between the independent and dependent variables
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R^2 Score:", r2_score(y_test, y_pred))
```
